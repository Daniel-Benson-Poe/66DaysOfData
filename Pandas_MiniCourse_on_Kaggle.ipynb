###The DataFrame

A dataframe is a table containing an array of entries with certain values. Every entry corresponds to a row and a column.

`````python
pd.DataFrame(
    {"Yes": [50, 21], 
     "No" : [131, 2]}
)
`````

DataFrames can also contain strings as values instead

`````python
pd.DataFrame(
    {"Bob" : ["I liked it.", "It was awful."], 
     "Sue" : ["Pretty good.", "Bland."]}
)
`````

Declaration syntax for DataFrames is a dictionary having its keys as the column names and having values as the list of entries

The default when creating a DataBase is to have the row labels simply (0, 1, 2...) but we can specify how we want the rows labeled if we want

`````python
pd.DataFrame(
    {"Bob" : ["I liked it.", "It was awful."], 
     "Sue" : ["Pretty good.", "Bland."]},
     index=["Product A", "Product B"]
)
`````

###Series

A series is a sequence of data values. A Series is to a list as a DataFrame is to a table

`````python
pd.Series([1, 2, 3, 4, 5])
`````

A Series is a single column of a DataFrame. This means we can assign row labels to the Series using an index but the Series does not have an actual column name, only its overall name.

`````python
pd.Series(
    [30, 35, 40],
    index=["2015 Sales", "2016 Sales", "2017 Sales"],
    name="Product A"
)
`````

Excerpt from Kaggle:

"When choosing or transitioning between loc and iloc, there is one "gotcha" worth keeping in mind, which is that the two methods use slightly different indexing schemes.

iloc uses the Python stdlib indexing scheme, where the first element of the range is included and the last one excluded. So 0:10 will select entries 0,...,9. loc, meanwhile, indexes inclusively. So 0:10 will select entries 0,...,10.

Why the change? Remember that loc can index any stdlib type: strings, for example. If we have a DataFrame with index values Apples, ..., Potatoes, ..., and we want to select "all the alphabetical fruit choices between Apples and Potatoes", then it's a lot more convenient to index df.loc['Apples':'Potatoes'] than it is to index something like df.loc['Apples', 'Potatoet'] (t coming after s in the alphabet).

This is particularly confusing when the DataFrame index is a simple numerical list, e.g. 0,...,1000. In this case df.iloc[0:1000] will return 1000 entries, while df.loc[0:1000] return 1001 of them! To get 1000 elements using loc, you will need to go one lower and ask for df.loc[0:999].

Otherwise, the semantics of using loc are the same as those for iloc."

###Maps

A map takes a set of values and "maps" them to another set of values. There are two methods available: map() and apply().

`````python
# map() is used for a Series
# If we wanted to remean the scores the wines recieved to 0:
review_points_mean = reviews.points.mean()
reviews.points.map(lambda p: p - review_points_mean)
`````

`````python
# map() returns a new Series where all values have been transformed by your function
`````

`````python
# apply() is equivalent for transforming an entire DataFrame
def remean_points(row):
  row.points = row.points - review_points_mean
  return row

reviews.apply(remean_points, axis='columns')
`````

In the above, if instead of inputting axis='columns' we had put axis='index' we would be ransforming each column instead of each row.

Keep in mind that map() and apply() do not change the original Series or DataFrame, they simply return a NEW Series or DataFrame with the changes made. The originals are still the same

`````python
# However, pandas allows us a faster way to remean our points column
review_points_mean = reviews.points.mean()
reviews.points - review_points_mean
`````

Pandas looks at the above code and determines that our meaning is to subtract the mean value from every value in the dataset

`````python
# Pandas can also combine country and region in the dataset; it knows to perform the operations between Series of equal length
reviews.country + " - " + reviews.region_1
`````

The built-in pandas operators are acctually faster than map() and apply() due to built-in speed ups, but they are not as flexible as map() and apply().

We'd like to host these wine reviews on our website, but a rating system ranging from 80 to 100 points is too hard to understand - we'd like to translate them into simple star ratings. A score of 95 or higher counts as 3 stars, a score of at least 85 but less than 95 is 2 stars. Any other score is 1 star.

Also, the Canadian Vintners Association bought a lot of ads on the site, so any wines from Canada should automatically get 3 stars, regardless of points.

Create a series star_ratings with the number of stars corresponding to each review in the dataset.

`````python
reviews.columns
`````

`````python
reviews['points'].describe()
`````

`````python
def fix_points(row):
  if row.country == 'Canada':
    return 3
  elif row.points >= 95:
    return 3
  elif row.points >= 85:
    return 2
  else:
    return 1
`````

`````python
star_ratings = reviews.apply(fix_points, axis='columns')
`````

#Grouping and Sorting

`````python
# Initial Setup
import pandas as pd
reviews = pd.read_csv("/content/winemag-data-130k-v2.csv", index_col=0)
pd.set_option("display.max_rows", 5)
`````

`````python
# We can groupby to sort the data
reviews.groupby("points").points.count()
`````

`````python
# We can get the cheapest wine in each point value cat.
reviews.groupby('points').price.min()
`````

`````python
# We can select the name of the first wine reviewed from each winery
reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
`````

`````python
# We can also group by more than one column
reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
`````

`````python
# We can use agg() to run different functions on DataFrame simultaneously
reviews.groupby(['country']).price.agg([len, min, max])
`````

###Multi Indexing

`````python
# Sometimes the groupby will give us a multiindex
countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
`````

`````python
mi = countries_reviewed.index
type(mi)
`````

`````python
# We can convert a multi index back to a regular index
countries_reviewed.reset_index()
`````

###Sorting

`````python
# We can sort the data to get the order we want ourselves using sort_values()
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
`````

`````python
# sort_values() defaults to ascending sort so we can manually force the return to be descending
countries_reviewed.sort_values(by='len', ascending=False)
`````

`````python
# We can sort by index using sort_index()
countries_reviewed.sort_index()
`````

`````python
# We can even sort by more than one column at a time!
countries_reviewed.sort_values(by=['country', 'len'])
`````
