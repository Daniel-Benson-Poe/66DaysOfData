```python
#Create simple scatter plot with horizontal x-axis set to bmi and vertical 
# y-axis set to charges
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])

plt.show()
```

The scatterplot above shows that bmi and insurance charges look to be positively correlated. We can add a regression line to check the strength of this correlation. The regression line is the line that best fits the data.

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])

plt.show()
```

We can use scatterplots to see relationships between three different variables by using color coded scatter plots.

Let's look at how smoking affects BMI and insurance costs

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])

plt.show()
```

We see that nonsmokers tend to pay slightly more with increasing BMI, but smokers pay much more. 

Let's use a lmplot to add two regression lines.

```python
sns.lmplot(x='bmi', y='charges', hue="smoker", data=insurance_data)

plt.show()
```

We can also use a categorical scatter plot called a swarmplot to use a categorical variable on one of the main axes (where we would usually only use continuous variables)

```python
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])
```

We can see that on average non-smokers are charged less than smokers, and customers who pay the most are smokers and customers who pay the least are non-smokers.

#Choosing Plot Types and Custom Styles

##The Categories of Chart Types

###Trends

Trend is a pattern of change

sns.lineplot  Line charts; best to show trends over time with multiple lines available for use for trends in multiple groups.

###Relationship

Understanding relationships between variables within a dataset

 sns.barplot Bar charts; compare quantities corresponding to different groups

 sns.heatmap  Heatmaps; used for color-coded patterns in tables of numbers

 sns.scatterplot Scatter plots; relationships between two continuous variables or with color-coding a third categorical variable included.

 sns.regplot  regression line scatter plot; easier to see linear relationship between two variables

sns.lmplot  multiple regression lines in scatter plot with color-coded groups

sns.swarmplot  Categorical scatter plot; shows relationship between continuous variable and a categorical variable

###Distribution

Show possible values expected in a variable, along with how likely they are

 sns.histplot Histograms; show distribution of single numerical variable

 sns.kdeplot  KDE plots; show estimated, smooth distribution of single numerical variable (or two numerical variables  2d KDE)

 sns.jointplot  Simultaneously displaying 2D KDE plot with corresponding KDE plots for each individual variable
